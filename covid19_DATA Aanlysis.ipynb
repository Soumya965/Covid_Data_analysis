###  3.3: find maximum infection rate for all of the countries.

```python
countries = list(df_aggregated.index)
```

```python
countries
```

```python
max_infection_rates= []
for i in (countries):
    max_infection_rates.append(df_aggregated.loc[i].diff().max())
print(max_infection_rates)
```

```python
max(max_infection_rates)
```

```python
df_aggregated['max_infection_rates'] = max_infection_rates
```

```python
df_aggregated
```
